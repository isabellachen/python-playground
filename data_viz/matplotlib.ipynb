```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('./countries.csv')
```

```python
af = data[data.country == "Afghanistan"]
```

```python
plt.plot(af.year, af.gdpPerCapita)
plt.title("Afghanistan GDP")
```

```python
print(set(data.continent))
```

```python
data_2007 = data[data.year == 2007]
asia_2007 = data_2007[data_2007.continent == "Asia"]
europe_2007 = data_2007[data_2007.continent == "Europe"]
```

```python
countries_in_asia = len(set(asia_2007.country))
countries_in_europe = len(set(europe_2007.country))
print(countries_in_asia)
print(f"There are {countries_in_asia} in Asia")
print(countries_in_europe)
print(f"There are {countries_in_europe} in Europe")
```

```python
print(asia_2007.gdpPerCapita.median())
print(europe_2007.gdpPerCapita.median())
```

```python
plt.subplot(2,1,1)
plt.hist(asia_2007.gdpPerCapita, edgecolor="black")
plt.subplot(2,1,2)
plt.hist(europe_2007.gdpPerCapita, edgecolor="black")
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(2,1,figsize=(16,5))
fig.suptitle('GDP per capita')
ax1.hist(asia_2007.gdpPerCapita, range=(0, 50_000), edgecolor="black")
ax1.set_ylabel('Asia')
ax2.hist(europe_2007.gdpPerCapita, color="green", range=(0, 50_000), alpha=0.7, edgecolor="black")
ax2.set_ylabel('Europe')
```

```python
data_1997 = data[data.year == 2007]
eu_1997 = data_1997[data_1997.continent == "Europe"]
americas_1997 = data_1997[data_1997.continent == "Americas"]
```

```python
fig, (ax1, ax2) = plt.subplots(2,1, figsize=(16,5))
fig.suptitle('Life expectancy')
ax1.hist(eu_1997.lifeExpectancy, range=(55, 90))
ax2.hist(americas_1997.lifeExpectancy, range=(55, 90), color="green")
```
