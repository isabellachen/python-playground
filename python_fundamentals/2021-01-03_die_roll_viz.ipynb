```python
import matplotlib.pyplot as plt
import numpy as np
import random
import seaborn as sns
```

```python
rolls = [random.randrange(1,7) for i in range(600)]
values, freqs = np.unique(rolls, return_counts=True)
print(values, freqs)
```

```python
sns.set_style('whitegrid')
axes = sns.barplot(x=values, y=freqs, palette='bright')
axes.set_title('Dice Rolls')
plt.show()
```

```python
def draw_dice_plots(n=600):
    rolls = [random.randrange(1,7) for i in range(600)]
    values, freqs = np.unique(rolls, return_counts = True)
    
    fig, ax = plt.subplots()
    ax.set_ylim(top=max(freqs) * 1.20)
    sns.set_style('whitegrid')
    ax.set_xlabel('Die Faces')
    ax.set_ylabel('Die Freqs')
    sns.barplot(x=values, y=freqs, ax=ax)
    
    # ax.patches are bars in the bar plot
    for bar, freq in zip(ax.patches, freqs):
        text_x = bar.get_x() + bar.get_width()/2.0
        text_y = bar.get_height()
        text = f'{freq:,}\n{freq/len(rolls):.3%}'
        ax.text(text_x, text_y, text, fontsize=11, ha='center', va='bottom')
    
print(draw_dice_plots())
```
